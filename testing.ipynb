```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(
    np.random.randint(10, 100, (10, 2)),
    columns=["A", "B"],
    index=[f"idx_{i}" for i in range(1, 11)],
)
```

```python
df.head()
```

```python
sns.lineplot(data=df, x=df.index, y="B", label="B", markers=True, style=["A"] * len(df))
sns.lineplot(data=df, x=df.index, y="A", label="A", markers=True, style=["B"] * len(df))
# plt.legend()
```
